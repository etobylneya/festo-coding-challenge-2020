```python
df = pd.read_csv(r'2_2_christmas_shopping.csv', delimiter = ', ', engine = 'python')

row_1 = [num for num in df.columns]
row_1[9] = 22
row_1[6] = 18

df.columns = ['work', 's1', 's2', 'p1', 'p2', 'h1', 'h2', 'd1', 'd2', 't1', 't2', 'home']

df.loc[-1] = row_1
df.index = df.index + 1
df = df.sort_index()

df['start'] = ['work', 's1', 's2', 'p1', 'p2', 'h1', 'h2', 'd1', 'd2', 't1', 't2', 'home']
df = df.set_index('start')

df.drop('home', axis = 0, inplace = True)
df
```

```python
from itertools import product, permutations
```

```python
combinations = ['s', 'p', 'h', 'd', 't']
lst_1 = ['s1', 'p1', 'h1', 'd1', 't1', 'home']
lst_2 = ['s2', 'p2', 'h2', 'd2', 't2', 'home']
min_length = np.inf
optimal_symbols = 's1p1h1d1t1'
for combination in permutations(combinations, 5):

    combination = list(combination)

    for option in product(['1', '2'], repeat = 5):
        option = ''.join(option)

        route = ['', '', '', '', '']
        for i in range(5):
            if option[i] == '1':
                route[i] = lst_1[combinations.index(combination[i])]
                
            elif option[i] == '2':
                route[i] = lst_2[combinations.index(combination[i])]        
                 
        length = 0
        for i in range(len(route)):
            if i == 0:
                length += int(df[route[i]].loc['work'])
            else:
                length += int(df[route[i]].loc[route[i - 1]])
        length += int(df['home'].loc[route[-1]])
        if length < min_length:
            min_length = length
            str = ''
            for symbol in route:
                str += symbol
            optimal_symbols = str

print(f'{optimal_symbols} - {min_length}')
            
```

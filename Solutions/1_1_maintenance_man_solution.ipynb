# 7(1.1) FIVE XOR SEVEN

```python
rooms = []
for number in range(2500):
    number = str(number)
    cntr_5 = number.count('5')
    cntr_7 = number.count('7')
    if ((cntr_5 >= 1) and (cntr_7 == 0)) or ((cntr_5 == 0) and (cntr_7 >= 1)):
        rooms.append(number)
print(f'1000th room number is {rooms[999]}')
```

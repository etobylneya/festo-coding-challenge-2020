The data is slightly off comparing to the table on the challenge page, so we need to make it look right before we can start solving the challenge.

```python
df = pd.read_csv(r'1_2_christmas_cards.csv', delimiter = ', ', engine = 'python')
row_1 = [num for num in df.columns]
row_1[7] = 9

df.columns = ['work', 'a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'home']

df.loc[-1] = row_1
df.index = df.index + 1
df = df.sort_index()
df['start'] = ['work', 'a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'home']
df = df.set_index('start')

df = df.drop(['home'], axis = 0)
df
```

This is what it is supposed to look like. Way better!

```python
from itertools import permutations
```

```python
lst = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h']

min_length = 100
optimal_symbols = 'abcdefgh'
# for every possible combination of friends lets calculate the length of the route and print the shortest
for combination in permutations(lst, 8):
    length = 0
#     combination variable is a list and I need it to be a string
    combination = ''.join(combination)
#     go from 0 to 7 and sum the distance between friends
    for i in range(len(combination)):
#         if this is the 1st friend in the string (0th to be precise) let's get the distance from work to the friend
        if i == 0:
            length += int(df[combination[i]].loc['work']) 
#         otherwise lets get the distance from previous friend to the current friend
        else:
            length += int(df[combination[i]].loc[combination[i - 1]])
#     after the loop lets add the distance from last friend to home
    length += int(df['home'].loc[combination[-1]])
#     if current combination gives us the shortest route yet, let's store it as the optimal
    if length < min_length:
        min_length = length
        optimal_symbols = combination
        
print(f'optimal route is {optimal_symbols}')
```

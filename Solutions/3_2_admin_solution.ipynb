```python
import pandas as pd
```

```python
df = pd.read_csv(r'3_2_delivery_service.csv', delimiter = ', ', engine = 'python')
row_1 = ['-', '6', '3', '22', '20', '10', '-', '-', '-', '-', '-', '-']
df.columns = ['work', 'r1', 'r2', 'r3', 'r4', 'r5', 'c1', 'c2', 'c3', 'c4', 'c5', 'home']
# df.loc[0] = row_1
df.loc[-1] = row_1
df.index = df.index + 1
df = df.sort_index()
df['start'] = ['work', 'r1', 'r2', 'r3', 'r4', 'r5', 'c1', 'c2', 'c3', 'c4', 'c5', 'home']
df = df.set_index('start')
df.drop(['home'], inplace = True)
df
```

```python
from itertools import permutations
import numpy as np
```

```python
combinations = ['r1', 'r2', 'r3', 'r4', 'r5', 'c1', 'c2', 'c3', 'c4', 'c5']
restaurants = ['r1', 'r2', 'r3', 'r4', 'r5']
customers = ['c1', 'c2', 'c3', 'c4', 'c5']

optimal_path = ''
min_length = np.inf

for combination in permutations(combinations):
    combination_str = ''.join(combination)

    not_acceptible = False
    for i in range(5):
        r_index = combination_str.find(restaurants[i])
        c_index = combination_str.find(customers[i])
        
        if r_index > c_index:
            not_acceptible = True
    if not_acceptible:
        continue
        
    num_meals = 0
    combination_lst = list(combination)

    for i in range(len(combination)):

        if combination_lst[i] in restaurants:
            num_meals += 1
            
        elif combination_lst[i] in customers:
            num_meals -= 1
            
        if num_meals > 3:
            not_acceptible = True
    
    if not_acceptible:
        continue

    length = 0
    for i in range(len(combination_lst)):
        if i == 0:
            length = int(df[combination_lst[i]].loc['work'])
            
        else:
            length += int(df[combination_lst[i]].loc[combination_lst[i - 1]])
            
    length += int(df['home'].loc[combination_lst[-1]])
    
    if length < min_length:
        min_length = length
        optimal_path = ''        
        for symbol in combination_lst:
            optimal_path += symbol
            
print(optimal_path, min_length)
```
